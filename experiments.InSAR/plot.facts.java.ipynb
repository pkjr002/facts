```python
import pandas as pd
from io import StringIO
#
InsarGPS_FILE='javaIsland_InSAR_rate.txt'
InsarGPS_FILEpath=f'/Users/pk695/werk.M2/FACTS_dev/202411_java.java/{InsarGPS_FILE}'
#
with open(InsarGPS_FILEpath, 'r') as f:
    lns = f.readlines()
col_nme = lns[0].strip().split(', ')
data    = "\n".join(lne.strip() for lne in lns[1:])
df      = pd.read_csv(StringIO(data), delim_whitespace=True, names=col_nme, dtype=str)
array   = df.to_numpy()
# this contains string to preserve the 0's
idLatLon_tar_str=array[:,0:3]
idLatLon_tar = array[:,0:3].astype(float)
```

```python
facts_newloc_FOLDER=f'./'
facts_newloc_lstFILE='latlon_basedON_inDATA_.txt'
#
# Write the location.lst file.
longitude = idLatLon_tar_str[:,0]
latitude = idLatLon_tar_str[:,1]
count = 0   
with open(f'{facts_newloc_FOLDER}/{facts_newloc_lstFILE}', "w") as file:
    for i, (lat, lon) in enumerate(zip(latitude, longitude)):
        #
        grid_lat_lon = f"grid_{lat}_{lon}"
        file.write(f"{grid_lat_lon}\t{i}\t{lat}\t{lon}\n")
        #
        if i == len(latitude) - 1: break 
```
