```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import netCDF4 as nc
from pandas.plotting import table 
import xarray as xr
from datetime import date
import os
#
import warnings
import matplotlib
warnings.filterwarnings("ignore", category=matplotlib.MatplotlibDeprecationWarning)
#
colrs = 'rcgmbyrkcbgm'
```

### <font color=blue> List of **Modules** and **Data** to be used </font>

```python
# Load module nc data.
#
df_module = pd.read_fwf('001_input_Names_Modules_dataFILES/Fig7_IP_module_submodule_DataFILE_local.txt',comment = '#')
df_module
```

## <font color=green> **Compute** and **Plot** <font color=black>*Fraction of variance*  </font>

```python
import fig_varDecomp as IP
#
EXPDATE  = 221217
wf1f = ['WAIS-ipccar5', 'EAIS-ipccar5' , 'GrIS-FittedISMIP' ,  'Glaciers-ipccar5-GMIP2'  ,  'LWS-ssp' , 'TE-tlm' , 'VLM-kopp14' ]
wf2f = ['AIS-larmip'    , 'GrIS-FittedISMIP' ,  'Glaciers-ipccar5-GMIP2'  ,  'LWS-ssp' , 'TE-tlm' , 'VLM-kopp14' ]
wf3f = ['AIS-deconto21' , 'GrIS-FittedISMIP' ,  'Glaciers-ipccar5-GMIP2'  ,  'LWS-ssp' , 'TE-tlm' , 'VLM-kopp14' ]
wf4  = ['AIS-bamber19'  , 'GrIS-bamber19'    ,  'Glaciers-ipccar5-GMIP2'  ,  'LWS-ssp' , 'TE-tlm' , 'VLM-kopp14' ]
#
workflow = ['wf1f', 'wf2f', 'wf3f','wf4']
ssp      = ['ssp126','ssp585'] 
#
# ---------  
#
df = df_module
#
plt.subplots(figsize=(10, 10));
# Subplot will alternate between ssp
subPLT = 0
#
for wf in enumerate(workflow):
    for ss in enumerate(ssp):
        #
        # Import the .nc data for each workflow's component/modules
        [sampsloccomponents, components, yrs] = IP.module_varFig_nc(df,eval(f'{wf[1]}'),[ss[1]],EXPDATE)
        #
        # wf1f doesnot have AIS, but has WAIS & EAIS. Add them.
        if wf[1] == 'wf1f' and (ss[1] == 'ssp126' or ss[1] == 'ssp585'):
            sampsloccomponents = np.concatenate((sampsloccomponents[:, :1, :] + sampsloccomponents[:, 1:2, :], sampsloccomponents[:, 2:, :]), axis=1)
            components                = ['AIS', 'GrIS' ,  'Glaciers'  ,  'LWS' , 'TE' , 'VLM' ]
        #
        #
        # Compute the Fraction of variance.
        [varV, varF] = IP.varV_varF(sampsloccomponents,components)
        # Since it is local, we use SD instead of TE.
        components[4] = 'SD';
        #
        # Plot Fraction of variance of workflow in subplot for 2 SSP.
        subPLT += 1
        ax = plt.subplot(4, 2, subPLT); ax.stackplot(yrs,varF[0],varF[1], varF[2], varF[3], varF[4], varF[5],labels=components, colors=colrs);
        #
        # Set same Y-axis limits across each workflow.
        if wf[1] == 'wf1f' and ss[1] == 'ssp126':  ax.set_ylim([0, 1]);  ax.set_title('New York: SSP1-2.6'); 
        if wf[1] == 'wf1f' and ss[1] == 'ssp585':  ax.set_ylim([0, 1]);  ax.set_title('New York: SSP5-8.5'); ax.legend(loc='lower left');
        #
        if wf[1] == 'wf2f' and ss[1] == 'ssp126':  ax.set_ylim([0, 1.05]); ax.axhline(y=1,color='black')  
        if wf[1] == 'wf2f' and ss[1] == 'ssp585':  ax.set_ylim([0, 1.05]); ax.axhline(y=1,color='black')
        #
        if wf[1] == 'wf3f' and ss[1] == 'ssp126':  ax.set_ylim([0, 1]);
        if wf[1] == 'wf3f' and ss[1] == 'ssp585':  ax.set_ylim([0, 1]);
        #
        if wf[1] == 'wf4'  and ss[1] == 'ssp126':  ax.set_ylim([0, ax.get_ylim()[1]+0.02]); ax.axhline(y=1,color='black')
        if wf[1] == 'wf4'  and ss[1] == 'ssp585':  ax.set_ylim([0, ax.get_ylim()[1]+0.02]); ax.axhline(y=1,color='black')
        #
        if wf[1] != 'wf4': ax.set_xlabel(''); ax.xaxis.set_ticklabels([])
        #
        #
        ax.set_xlim([2020, 2100]); 
        ax.set_ylabel('Fraction of variance');
        ax.text(.7, .1, 'Workflow '+wf[1][2:], horizontalalignment='left',verticalalignment='top',weight='bold',fontsize=10,transform=ax.transAxes)
        ax.tick_params(axis='both', labelsize=10); ax.yaxis.set_ticks_position('both')
        #
plt.subplots_adjust(wspace=0.35, hspace=0.4)
#
# Get todays date.
today = date.today().strftime('%Y-%m-%d')
# Save Figure.
figureNAME = "Fig7_VarDecomp_local"+today+".pdf"
if os.path.exists(figureNAME): os.remove(figureNAME)
plt.savefig(figureNAME, format="pdf", bbox_inches="tight")
# 
plt.show()
```
