```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#
import matplotlib.cm as cm
from matplotlib.colors import Normalize
from matplotlib.colorbar import ColorbarBase


ax = sns.kdeplot(x=d16_2020, y=d16_2100, fill=True, cmap="Blues", cbar=False)
# Create a color map and normalize
cmap1 = cm.Blues
norm1 = Normalize(vmin=min(d16_2020.min(), d16_2100.min()), vmax=max(d16_2020.max(), d16_2100.max()))
#
# Create an axis for the first color bar
cbar_ax1 = ax.figure.add_axes([0.92, 0.15, 0.02, 0.7])
# Create a color bar for the first plot
sm1 = cm.ScalarMappable(cmap=cmap1, norm=norm1)
sm1.set_array([])
cbar1 = plt.colorbar(sm1, cax=cbar_ax1)



ax = sns.kdeplot(x=k14_2020, y=k14_2300, fill=True, cmap="Reds", cbar=False)
# Create a color map and normalize
cmap2 = cm.Reds
norm2 = Normalize(vmin=min(k14_2020.min(), k14_2300.min()), vmax=max(k14_2020.max(), k14_2300.max()))
#
# Create an axis for the second color bar
cbar_ax2 = ax.figure.add_axes([1.0, 0.15, 0.02, 0.7])
# Create a color bar for the second plot
sm2 = cm.ScalarMappable(cmap=cmap2, norm=norm2)
sm2.set_array([])
cbar2 = plt.colorbar(sm2, cax=cbar_ax2)


plt.title("AIS contribution to GMSL in 2020 \n and that in 2100")
plt.xlabel("AIS contribution in 2020 (cm)")
plt.ylabel("AIS contribution in 2100 (cm)")
plt.show()


plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()
```
