```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from matplotlib.colors import LogNorm

# Sample data
# d16_2020 = np.random.exponential(scale=1.0, size=1000)
# d16_2300 = np.random.exponential(scale=1.0, size=1000)

# Calculate the KDE
x, y = np.mgrid[d16_2020.min():d16_2020.max():100j, d16_2300.min():d16_2300.max():100j]
positions = np.vstack([x.ravel(), y.ravel()])
values = np.vstack([d16_2020, d16_2300])
kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(positions).T, x.shape)

# Create contourf plot with logarithmic color scale
fig, ax = plt.subplots()
cset = ax.contourf(x, y, Z, cmap="Blues", norm=LogNorm(vmin=Z.min(), vmax=Z.max()))

# Adding color bar
cbar = plt.colorbar(cset, ax=ax)
cbar.set_label('Density')

# Setting the title
ax.set_title("AIS contribution to GMSL in 2020 and that in 2100", fontsize=6)

plt.show()
```
