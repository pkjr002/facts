<font color=blue>  Extract all the GPD variables from the NC files</font>

```python

EXPDATE  = 221217
SCENARIO = ['ssp126','ssp585']
WORKFLO  = ["wf1f","wf2f","wf3f","wf4"]
year     = ['2050','2100']
#
today      = date.today().strftime('%Y-%m-%d')
figureNAME = "Fig8_ESL_NC_info-"+today 
#
if os.path.exists(figureNAME): os.remove(figureNAME)
file = open(figureNAME, "w")
#
df_TEMP = []
for scenario in SCENARIO:
    for wf in WORKFLO:
            for yy in year:
                #
                d_nc = xr.open_dataset('ESL_nc_data/' + 'coupling.{arg1}.extremesealevel.extremesealevel.pointsoverthreshold.{arg2}_id12_year{arg3}_extremesl.nc'.format(arg1=scenario,arg2=wf,arg3=yy))
                #
                a_mhhw          = np.squeeze(np.around(d_nc.mhhw.values, decimals=4))
                a_mhhw_freq     = np.squeeze(np.around(d_nc.mhhw_freq.values, decimals=4))
                a_gpd_location  = np.squeeze(np.around(d_nc.gpd_location.values, decimals=4))
                a_gpd_scale     = np.around(np.median(d_nc.gpd_scale.values), decimals=4)
                a_gpd_shape     = np.around(np.median(d_nc.gpd_shape.values), decimals=4)
                a_prj_freq      = np.around(np.median(d_nc.projected_frequencies[49,:].values), decimals=8)
                a_his_freq      = np.around(np.median(d_nc.historical_frequencies[49,:].values), decimals=8)
                #
                a_quantiles     = d_nc.coords['quantiles'].values
                a_height        = d_nc.coords['heights'].values
                #
                #
                file.write( 'SCENARIO = '+scenario+ ' || WORKFLOW ='+wf + ' || YEAR = '+yy + '\n' + \
                        'mhhw_freq = '+ str(a_mhhw_freq) + '\n' + 'mhhw = '+str(a_mhhw) + '\n' + \
                        'gpd_location = '+ str(a_gpd_location) + '\n' + 
                        'gpd_scale = ' + str(np.median(a_gpd_scale )) + '\n' + \
                        'gpd_shape = ' + str(np.median(a_gpd_shape)) +'\n'+ \
                        '-------------'+'\n'+'\n' )
                df_TEMP.append( [wf,scenario,yy, a_mhhw, a_mhhw_freq,  a_gpd_location, a_gpd_scale, a_gpd_shape, a_prj_freq, a_his_freq] )
                #
df_NC_Deets  = pd.DataFrame( df_TEMP, columns=['Workflow','SSP','Year', 'MHHW', 'MHHW_FreQ','gpd_location', 'gpd_scale(50)', 'gpd_shape(50)', 'a_prj_freq(50)', 'a_his_freq(50)'])
df_NC_Deets1 = pd.DataFrame( df_NC_Deets.set_index( ['Workflow','Year','SSP']) ).unstack()#.swaplevel( 0,1, axis=1 )
display(df_NC_Deets1)
file.close()

                
```
