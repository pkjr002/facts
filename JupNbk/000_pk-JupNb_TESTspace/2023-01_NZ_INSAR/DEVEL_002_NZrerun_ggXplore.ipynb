```python
import os

def list_files(path, indent=0, file=None):
    for item in os.listdir(path):
        if os.path.isfile(os.path.join(path, item)):
            file.write(' ' * indent + '- ' + item + '\n')
            print(' ' * indent + '- ' + item)
        elif os.path.isdir(os.path.join(path, item)):
            file.write(' ' * indent + '+ ' + item + '\n')
            print(' ' * indent + '+ ' + item)
            list_files(os.path.join(path, item), indent + 2, file)

with open('directory_listing.txt', 'w') as file:
    list_files('/projects/kopp/ggg46/nz_projections/', file=file)
```
