```python
# Prep the RFF emissions
def prep_rff(baseem, rffemissions, rff_sp,REFERENCE_YEAR):
    
    # maps GHG gas to index in emissions numpy array
    idxdt = {
        "C": 1,
        "CH4": 3,
        "N2O": 4,
        "N2": 4,
    }
    styear = rffemissions.Year.values[0]
    enyear = rffemissions.Year.values[-1]

    # put the RFF-SP gases into the given background emissions 
    rffem_base  = baseem
    rffem_full  = baseem

    for gas in rffemissions.gas.values:
        # print(gas)
    #    
        rffem_full[styear-REFERENCE_YEAR:enyear-REFERENCE_YEAR+1,idxdt[gas]] = rffemissions.sel(gas=gas,rff_sp=rff_sp).emissions.values
    #    # rffem_full[styear-REFERENCE_YEAR:enyear-REFERENCE_YEAR+1,idxdt[gas]] = np.zeros(551 - 270)

    return rffem_base, rffem_full
```
